```python
def bench_absl(n=2**20, q=2**20):
    res = !./bazel-bin/absl-btree {n} {q}
    duration = float(res[0].split()[0])
    check = int(res[1])
    print(f'{n:<7} {q:<7} {check:10} {duration}')
    return duration
```

```python
bench_absl()
```

```python
ns = list(int(1.17**k) for k in range(30, 110))
absl = [bench_absl(n=n, q=2**22) for n in ns]
```

```python
import pickle

with open('absl.pkl', 'wb') as file:
    pickle.dump(absl, file)
```
