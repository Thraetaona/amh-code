```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
fnames = ['std', 'std-fat', 'aos', 'ska', 'linear', 'robin-hood', 'swiss1', 'swiss'][3:]

plt.figure(figsize=(16, 9), dpi=80)

for run in fnames:
    text = !cat {run}.run
    xs = []
    ys = []
    for line in text:
        x, y = line.split('\t')
        xs.append(int(x))
        ys.append(float(y))
    plt.plot(xs, ys)

plt.legend(fnames)
plt.xscale('log')
plt.ylim(0)
plt.show()
```
